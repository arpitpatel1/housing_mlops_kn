```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
path = Path('../data/interim')
```

```python
flats = pd.read_csv(path / 'flats_cleaned.csv')
houses = pd.read_csv(path / 'house_cleaned.csv')
```

```python
houses.head(5)
```

```python
df = pd.concat([flats,houses],ignore_index=True)
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df.sample(df.shape[0],ignore_index=True)
```

```python
df.head()
```

```python
df.to_csv(path / 'properties.csv',index=False)
```
